```python
import doltcli as dolt
import pandas as pd
```

```python
df = pd.read_csv('tasks_2023_06_10.csv', dtype={'ccn': str})
df
```

```python
db = dolt.Dolt("/root/data/transparency-in-pricing")
db
```

```python
def lookup_name_by_ccn(ccn):
    sql = "SELECT name FROM hospital WHERE id = '{}';".format(ccn)
    
    res = db.sql(sql, result_format="json")
    if res.get('rows') is None or len(res['rows']) != 1:
        return None

    return res['rows'][0]['name']

df['name'] = df['ccn'].apply(lookup_name_by_ccn)
df
```

```python
df[df['name'].isnull()]['transparency_page']
```

```python
df = df[df['name'].notnull()]
```

```python
df2 = pd.read_csv('tasks_2023_06_12.csv', dtype={'ccn': str})
df2['name'] = df2['ccn'].apply(lookup_name_by_ccn)
df2
```

```python
df2[df2['name'].isnull()]['transparency_page']
```

```python
df3 = pd.read_csv('tasks_adventist.csv', dtype={'ccn': str})
df3
```

```python
df3['name'] = df3['ccn'].apply(lookup_name_by_ccn)
df3
```

```python
df3[df3['name'].isnull()]
```

```python
df_final = pd.concat([df, df2, df3])
df_final
```

```python
df_final.to_csv('tasks_2023_06_13.csv', index=False)
```
