```python
def fix_header(df):
    orig_colnames = df.columns.to_list()
    new_colnames = df.iloc[0].to_list()

    columns = []
    
    for i in range(len(orig_colnames)):
        orig_colname = orig_colnames[i]
        new_colname = new_colnames[i]

        if type(new_colname) != str:
            columns.append(orig_colname)
            continue

        if orig_colname.startswith('Unnamed') or orig_colname in [ 'IP Expected Reimbursement', 'OP Expected Reimbursement', 'WVH' ]:
            columns.append(new_colname.replace('\n', ' ').strip())
        else:
            columns.append(orig_colname)

    df.columns = columns
    df = df.drop(df.index[[0,1]])

    return df

df_proc = fix_header(df_proc)
df_ms_drgs = fix_header(df_ms_drgs)
df_apr_drg = fix_header(df_apr_drg)
df_eapg = fix_header(df_eapg)
df_apc = fix_header(df_apc)
df_sup = fix_header(df_sup)
df_pharm = fix_header(df_pharm)
df_nonf = fix_header(df_nonf)
df_fac = fix_header(df_fac)
df_fac
```

```python
def fix_codes(row):
    def match_and_set(row, code):
        if code is None:
            return

        if code.startswith('Custom'):
            return
        
        if code.startswith('HCPCS') or code.startswith('CPT'):
            code = code.split(" ")[-1]
            
        if code_is_cpt(code) or code_is_hcpcs(code):
            row["hcpcs_cpt"] = code
            return

        if code.startswith('MS-DRG'):
            code = code.split(' ')[-1]
        
        if code_is_ms_drg(code):
            if row["ms_drg"] is None:
                row["ms_drg"] = code
        
    code = row.get("code")
    match_and_set(row, code)

    return row

df_out = df_out.apply(fix_codes, axis=1)
```

```python
def payer_name_to_rate_category(payer_name):
    if payer_name == "Cash":
        return "cash"
    elif payer_name == "OP Gross Charge" or payer_name == "IP Gross Charge" or payer_name == "GrossCharge":
        return "gross"
    elif payer_name == "Max" or payer_name == "Maximum":
        return "max"
    elif payer_name == "Min" or payer_name == "Minimum":
        return "min"

    return "payer"

df_out['rate_category'] = df_out['payer_name'].apply(payer_name_to_rate_category)
df_out.loc[df_out['payer_name'].str.startswith('OP '), 'setting'] = 'outpatient'
df_out.loc[df_out['payer_name'].str.startswith('IP '), 'setting'] = 'inpatient'
df_out
```
