```python
remaining_columns = df_mid.columns[:5].to_list()
df_mid = pd.melt(df_mid, id_vars=remaining_columns, var_name='payer_name2', value_name='standard_charge')
df_mid
```

```python
def payer_category_from_payer_name(payer_name):
    if payer_name == "gross charge":
        return "gross"
    elif payer_name == "discounted cash price":
        return "cash"
    elif payer_name == "de-identified minimum negotiated charge":
        return "min"
    elif payer_name == "de-identified maximum negotiated charge":
        return "max"
    elif payer_name == "payer-specific negotiated charge":
        return "payer"

    return "payer"

df_mid['payer_category'] = df_mid['payer_name2'].apply(payer_category_from_payer_name)
df_mid
```

```python
df_mid.loc[df_mid['payer_category'] != 'payer', 'payer_name'] = df_mid[df_mid['payer_category'] != 'payer']['payer_name2']
del df_mid['payer_name2']

df_mid
```
