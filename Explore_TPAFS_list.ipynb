```python
import pandas as pd
import requests
```

```python
url = "https://raw.githubusercontent.com/TPAFS/transparency-data/main/price_transparency/hospitals/machine_readable_links.csv"
df = pd.read_csv(url)
df
```

```python
df['supplemental_url'].value_counts()[:20]
```

```python
df = df[df['file_size'].notnull()]
```

```python
df = df[df['file_size'] < 100 * 1024 * 1024]
df
```

```python
df = df.sort_values('file_size', ascending=False)
df
```

```python
df = df[['ccn', 'reporting_entity_name_common', 'machine_readable_url', 'machine_readable_page']]
df
```

```python
def check_mrf_url(row):
    try:
        resp = requests.head(row['machine_readable_url'], timeout=3.0)
        print(resp.url, resp.status_code)
        row['http_status_code'] = resp.status_code
        if resp.status_code == 200:
            row['content_length'] = resp.headers.get('Content-Length')
    except Exception as e:
        print(e)

    return row

df = df.apply(check_mrf_url, axis=1)
df
```

```python
df = df[df['http_status_code'].notnull()]
df['http_status_code'] = df['http_status_code'].astype(int)
df = df[df['http_status_code'] == 200]
df
```

```python
df = df[df['content_length'].notnull()]
df['content_length'] = df['content_length'].astype(int)
df = df[df['content_length'] < 150 * 1024 * 1024]
df
```

```python
df = df.sort_values('content_length', ascending=False)
df
```

```python
df.to_csv('todo.csv', index=False)
```
