```python
import pandas as pd
```

```python
TARGET_COLUMNS = [
    'hospital_id',
    #'row_id',
    'line_type',
    'description',
    'rev_code',
    'local_code',
    'code',
    'ms_drg',
    'apr_drg',
    'eapg',
    'hcpcs_cpt',
    'modifiers',
    'alt_hcpcs_cpt',
    'thru',
    'apc',
    'icd',
    'ndc',
    'drug_hcpcs_multiplier',
    'drug_quantity',
    'drug_unit_of_measurement',
    'drug_type_of_measurement',
    'billing_class',
    'setting',
    'payer_category',
    'payer_name',
    'plan_name',
    'standard_charge',
    'standard_charge_percent',
    'contracting_method',
    'additional_generic_notes',
    'additional_payer_specific_notes'
]
```

```python
# https://www.pennstatehealth.org/patients-visitors/billing-medical-records/understanding-cost-your-care
ccn = "399807"
url = "https://www.pennstatehealth.org/sites/default/files/Finance/November%202022/251854772_The%20Milton%20S.%20Hershey%20Medical%20Center_standardcharges.csv"
filename = url.split("/")[-1].replace("%20", " ")
ein = filename.split("_")[0]
ein = ein[:2] + "-" + ein[2:]
```

```python
import subprocess
subprocess.run(["wget", "--no-clobber", url, "-O", filename])
```

```python
df_in = pd.read_csv(filename, encoding='latin-1', low_memory=False)
df_in
```

```python
df_mid = pd.DataFrame(df_in)
df_mid = df_mid.rename(columns={
    'Line Type': 'line_type',
    'HMC CDM Code': 'local_code',
    'HMC CDM Description': 'description',
    'DRG': 'ms_drg',
    'CPT / HCPCS': 'hcpcs_cpt',
    'HMC Revenue Code': 'rev_code'
})

df_mid
```

```python
money_columns = df_mid.columns.to_list()[6:]
remaining_columns = df_mid.columns.to_list()[:6]
df_mid = pd.melt(df_mid, id_vars=remaining_columns, 
                 var_name='payer_name', value_name='standard_charge')
df_mid
```

```python
df_mid = df_mid.dropna(subset=['standard_charge'])
df_mid = df_mid[df_mid['standard_charge'] != " N/A "]

df_mid
```

```python
df_mid.loc[df_mid['line_type'] == 'DRG', 'setting'] = 'inpatient'
df_mid.loc[df_mid['hcpcs_cpt'].notnull(), 'setting'] = 'outpatient'
df_mid
```

```python
def payer_category_from_payer_orig(payer_orig):
    if payer_orig == "Gross or Average Charge":
        return 'gross'
    elif payer_orig == "Self Pay Cash Price":
        return 'cash'
    elif payer_orig == "Min Negotiated Rate":
        return 'min'
    elif payer_orig == "Max Negotiated Rate":
        return 'max'
    
    return 'payer'

df_mid['payer_category'] = df_mid['payer_name'].apply(payer_category_from_payer_orig)
df_mid
```

```python
def pad_drg_if_needed(drg):
    if drg is None:
        return 
    
    drg = str(drg).split(".")[0]
    if drg == "nan":
        return None
    
    if len(drg) == 1:
        return "00" + drg
    elif len(drg) == 2:
        return "0" + drg

    return drg

df_mid['ms_drg'] = df_mid['ms_drg'].apply(pad_drg_if_needed)
df_mid['ms_drg']
```

```python
df_mid['code'] = None
df_mid.loc[df_mid['line_type'] == 'DRG', 'code'] = df_mid['ms_drg']
df_mid.loc[df_mid['hcpcs_cpt'].notnull(), 'code'] = df_mid['hcpcs_cpt']

df_mid
```

```python
df_mid['apc'] = None
df_mid['apr_drg'] = None
df_mid['eapg'] = None
df_mid['alt_hcpcs_cpt'] = None
df_mid['thru'] = None
df_mid['icd'] = None
df_mid['drug_hcpcs_multiplier'] = None
df_mid['drug_quantity'] = None
df_mid['drug_unit_of_measurement'] = None
df_mid['drug_type_of_measurement'] = None
df_mid['billing_class'] = None
df_mid['plan_name'] = None
df_mid['modifiers'] = None
df_mid['standard_charge_percent'] = None
df_mid['contracting_method'] = None
df_mid['ndc'] = None
df_mid['additional_generic_notes'] = None
df_mid['additional_payer_specific_notes'] = None
df_mid['hospital_id'] = ccn

def pad_rev_code_if_needed(rev_code):
    if type(rev_code) == str and rev_code != 'na':
        if len(rev_code) == 3:
            return '0' + rev_code
        elif len(rev_code) == 2:
            return '00' + rev_code
        elif len(rev_code) == 1:
            return '000' + rev_code

    return rev_code

df_mid['rev_code'] = df_mid['rev_code'].apply(pad_rev_code_if_needed)
df_mid.loc[df_mid['rev_code'] == '0278<and>0278', 'rev_code'] = '0278'
df_mid
```

```python
df_out = pd.DataFrame(df_mid[TARGET_COLUMNS])
df_out = df_out.dropna(subset=['standard_charge'], axis=0)
df_out['standard_charge'] = df_out['standard_charge'].astype(str)
df_out['standard_charge'] = df_out['standard_charge'].apply(lambda rate: rate.replace(",", "").strip())
df_out = df_out[df_out['standard_charge'] != 'No Current Volume']
df_out = df_out[df_out['standard_charge'] != '-']
df_out
```

```python
df_out.to_csv(ccn + ".csv", index=False)
```
