```python
import pandas as pd
```

```python
TARGET_COLUMNS = [ 
    'hospital_ccn',
    'hospital_ein',
    'code_type',
    'internal_code',
    'description',
    'code_orig',
    'code_prefix',
    'code',
    'modifier',
    'ndc',
    'ndc_num_hcpcs_doses',
    'ndc_quantity_desc',
    'apc',
    'rev_code',
    'rev_desc',
    'billing_class',
    'patient_class',
    'payer_category',
    'payer_orig',
    'payer_name',
    'plan_orig',
    'plan_name',
    'plan_id',
    'plan_type',
    'rate',
    'rate_method',
    'rate_desc',
    'filename',
    'file_last_updated',
    'url',
    'permalink'
]
```

```python
ccn = "399807"
url = "https://www.pennstatehealth.org/sites/default/files/Finance/November%202022/251854772_The%20Milton%20S.%20Hershey%20Medical%20Center_standardcharges.csv"
filename = url.split("/")[-1].replace("%20", " ")
ein = filename.split("_")[0]
ein = ein[:2] + "-" + ein[2:]
```

```python
import subprocess
subprocess.run(["wget", "--no-clobber", url, "-O", filename])
```

```python
df_in = pd.read_csv(filename, encoding='latin-1', low_memory=False)
df_in
```

```python
df_mid = pd.DataFrame(df_in)
df_mid = df_mid.rename(columns={
    'Line Type': 'code_type',
    'HMC CDM Code': 'internal_code',
    'HMC CDM Description': 'description',
    'DRG': 'drg',
    'CPT / HCPCS': 'cpt',
    'HMC Revenue Code': 'rev_code'
})

df_mid
```

```python
money_columns = df_mid.columns.to_list()[6:]
remaining_columns = df_mid.columns.to_list()[:6]
df_mid = pd.melt(df_mid, id_vars=remaining_columns, var_name='payer_orig', value_name='rate')
df_mid
```

```python
df_mid = df_mid.dropna(subset=['rate'])
df_mid = df_mid[df_mid['rate'] != " N/A "]

df_mid
```

```python
df_mid.loc[df_mid['code_type'] == 'DRG', 'code_type'] = 'drg'
df_mid.loc[df_mid['cpt'].notnull(), 'code_type'] = 'cpt'
df_mid['code_prefix'] = 'none'
df_mid.loc[df_mid['code_type'] == 'drg', 'code_prefix'] = 'ms-drg'
df_mid.loc[df_mid['code_type'] == 'cpt', 'code_prefix'] = 'hcpcs_cpt'
df_mid['patient_class'] = 'na'
df_mid.loc[df_mid['code_type'] == 'cpt', 'patient_class'] = 'outpatient'
df_mid.loc[df_mid['code_type'] == 'drg', 'patient_class'] = 'inpatient'
df_mid
```

```python
def payer_category_from_payer_orig(payer_orig):
    if payer_orig == "Gross or Average Charge":
        return 'gross'
    elif payer_orig == "Self Pay Cash Price":
        return 'cash'
    elif payer_orig == "Min Negotiated Rate":
        return 'min'
    elif payer_orig == "Max Negotiated Rate":
        return 'max'
    
    return 'payer'

df_mid['payer_category'] = df_mid['payer_orig'].apply(payer_category_from_payer_orig)
df_mid
```

```python
df_mid['billing_class'] = 'na'
df_mid['plan_orig'] = 'na'
```

```python
def pad_drg_if_needed(drg):
    if drg is None:
        return 
    
    drg = str(drg).split(".")[0]
    if drg == "nan":
        return None
    
    if len(drg) == 1:
        return "00" + drg
    elif len(drg) == 2:
        return "0" + drg

    return drg

df_mid['drg'] = df_mid['drg'].apply(pad_drg_if_needed)
df_mid['drg']
```

```python
df_mid['code'] = 'na'
df_mid.loc[df_mid['code_type'] == 'drg', 'code'] = df_mid['drg']
df_mid.loc[df_mid['code_type'] == 'cpt', 'code'] = df_mid['cpt']
del df_mid['drg']
del df_mid['cpt']
df_mid
```

```python
df_mid['modifier'] = 'na'
df_mid['ndc'] = 'na'
df_mid['rev_code'] = df_mid['rev_code'].fillna('na')
df_mid['internal_code'] = df_mid['internal_code'].fillna('na')
df_mid['filename'] = filename
df_mid['url'] = url
df_mid['file_last_updated'] = "2022-06-30"
df_mid['hospital_ccn'] = ccn
df_mid['hospital_ein'] = ein

def pad_rev_code_if_needed(rev_code):
    if type(rev_code) == str and rev_code != 'na':
        if len(rev_code) == 3:
            return '0' + rev_code
        elif len(rev_code) == 2:
            return '00' + rev_code
        elif len(rev_code) == 1:
            return '000' + rev_code

    return rev_code

df_mid['rev_code'] = df_mid['rev_code'].apply(pad_rev_code_if_needed)
df_mid.loc[df_mid['rev_code'] == '0278<and>0278', 'rev_code'] = '0278'
```

```python
df_out = pd.DataFrame(columns=TARGET_COLUMNS)
df_out = df_out.append(df_mid)
df_out = df_out.dropna(subset=['rate'], axis=0)
df_out['rate'] = df_out['rate'].astype(str)
df_out['rate'] = df_out['rate'].apply(lambda rate: rate.replace(",", "").strip())
df_out = df_out[df_out['rate'] != 'No Current Volume']
df_out = df_out[df_out['rate'] != '-']
df_out
```

```python
df_out.to_csv(ccn + ".csv", index=False)
```
