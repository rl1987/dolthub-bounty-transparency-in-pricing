```python
from helpers import *
```

```python
TARGET_COLUMNS = [
    'hospital_tin', 
    'hospital_ccn', 
    'code_type',
    'description', 
    'procedure_code', 
    'code_orig',
    'code_prefix',
    'code', 
    'modifier', 
    'ndc', 
    'rev_code', 
    'rev_desc', 
    'apc', 
    'billing_class', 
    'patient_class', 
    'billed_quantity', 
    'quantity_desc', 
    'payer_orig', 
    'payer_category', 
    'payer_name', 
    'plan_orig', 
    'plan_name',
    'plan_id', 
    'plan_type', 
    'rate', 
    'rate_method', 
    'rate_desc',
    'filename',
    'file_last_updated', 
    'url', 
    'permalink'
]
```

```python
ccn = "100151"
url = "https://mcorgstatic.blob.core.windows.net/cms-price/590714831_mayo-clinic-florida_standardcharges.csv"
```

```python
filename = urlparse(url).path.split("/")[-1]
filename
```

```python
df_in = pd.read_csv(filename, low_memory=False)
df_in
```

```python
last_updated_at = df_in['code_desc'][1]
last_updated_at = last_updated_at.split(': ')[-1].replace('/', '-')
last_updated_at
```

```python
df_in = df_in.dropna(axis=0, subset=['rate_method'])
df_in
```

```python
df_out = pd.DataFrame(columns=TARGET_COLUMNS)
df_out
```

```python
df_in.columns
```

```python
df_mid = pd.DataFrame(df_in)
df_mid = df_mid.rename(columns={
    'code_desc': 'description',
    'setting': 'patient_class',
    'implied_quantity': 'quantity_type',
    'hcpcs_dose': 'quantity_desc',
    'contract_name': 'payer_orig',
    'payer_specific_negotiated_rate': 'rate',
    'minimum_negotiated': 'min',
    'maximum_negotiated': 'max',
    'standard_gross_charge': 'gross',
    'discounted_cash_price': 'cash',
})
df_mid
```

```python
money_columns = ['rate', 'min', 'max', 'gross', 'cash']
df_mid[money_columns] = df_mid[money_columns].apply(lambda values: cleanup_values(values))
del df_mid['quantity_type']
del df_mid['hcpcs']
del df_mid['dose_type']
del df_mid['applicable_percent']
del df_mid['count_of_compared_rates']
del df_mid['footnote']
df_mid
```

```python
columns = df_mid.columns.to_list()
remaining_cols = list(set(columns) - set(money_columns))
df_mid = pd.melt(df_mid, id_vars=remaining_cols)
df_mid = df_mid.rename(columns={
    'value': 'rate',
    'variable': 'payer_category'
})
df_mid.head(20)
```

```python
df_mid['patient_class'] = df_mid['patient_class'].fillna('na')
df_mid['patient_class'] = df_mid['patient_class'].apply(lambda pc: pc.lower())
df_mid['billing_class'] = df_mid['billing_class'].fillna('na')
df_mid['billing_class'] = df_mid['billing_class'].apply(lambda bc: bc.lower())
df_mid['billed_quantity'] = df_mid['billed_quantity'].fillna(-1)
df_mid['procedure_code'] = 'na'
df_mid['payer_category'] = df_mid['payer_category'].replace('rate', 'payer')
df_mid.sample(20)
```

```python
del df_mid['product_name']
df_mid['code_prefix'] = df_mid['code_type'].apply(lambda code_type: code_type.lower().replace('hcpcs', 'hcpcs_cpt').replace('ndc', 'none'))
df_mid['code_type'] = df_mid['code_type'].replace('CDM', 'cdm').replace('NDC', 'ndc').replace("HCPCS", 'cpt').replace("MS-DRG", 'drg')
df_mid['modifier'] = df_mid['modifier'].fillna('na')
df_mid['modifier'] = df_mid.apply(lambda row: row['modifier'][:7] if row['code_prefix'] == 'hcpcs_cpt' else 'na', axis=1)
df_mid['apc'] = 'na'
df_mid['ndc'] = df_mid.apply(lambda row: row['code'][:13] if row['code_type'] == 'ndc' else 'na', axis=1)
df_mid['ndc'] = df_mid['ndc'].apply(lambda ndc: ndc[:5] + '-' + ndc[5:] if len(ndc) == 9 else ndc)
df_mid['ndc'] = df_mid['ndc'].apply(lambda ndc: ndc + '  ' if len(ndc) == 8 else ndc)
df_mid['ndc'] = df_mid['ndc'].apply(lambda ndc: ndc + '   ' if len(ndc) == 7 else ndc)
df_mid['code'] = df_mid.apply(lambda row: 'na' if row['code_type'] == 'ndc' else row['code'], axis=1)
df_mid['code_type'] = df_mid.apply(lambda row: None if row['code_type'] == 'ndc' else row['code_type'], axis=1)
df_mid.sample(20)
```

```python
df_mid['url'] = url
df_mid['filename'] = filename
df_mid['file_last_updated'] = last_updated_at

hospital_tin = filename.split("_")[0]
hospital_tin = hospital_tin[:2] + "-" + hospital_tin[2:]
df_mid['hospital_tin'] = hospital_tin
df_mid['hospital_ccn'] = ccn
```

```python
df_mid['plan_orig'] = df_mid['payer_orig']
df_mid['plan_id'] = df_mid.apply(lambda row: row['payer_orig'].split("_")[0] if row['payer_category'] == 'payer' else None, axis=1)
df_mid['rev_desc'] = df_mid['rev_desc'][:100]
df_mid['rev_code'] = df_mid['rev_code'].fillna('na')
df_mid.sample(20)
```

```python
df_out = df_out.append(df_mid)
df_out.dropna(subset=['rate'], inplace=True)
df_out.sample(20)
```

```python
df_out.to_csv(ccn + ".csv", index=False)
```
