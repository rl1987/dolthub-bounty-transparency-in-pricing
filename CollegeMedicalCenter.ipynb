```python
import pandas as pd
TARGET_COLUMNS = [ 
    'hospital_ccn',
    'hospital_ein',
    'code_type',
    'internal_code',
    'description',
    'code_orig',
    'code_prefix',
    'code',
    'modifier',
    'ndc',
    'ndc_num_hcpcs_doses',
    'ndc_quantity_desc',
    'apc',
    'rev_code',
    'rev_desc',
    'billing_class',
    'patient_class',
    'payer_category',
    'payer_orig',
    'payer_name',
    'plan_orig',
    'plan_name',
    'plan_id',
    'plan_type',
    'rate',
    'rate_method',
    'rate_desc',
    'filename',
    'file_last_updated',
    'url',
    'permalink'
]
```

```python
ccn = "050776"
ein = "46-2760935"
url = "https://assets.changehealthcare.com/Shop/PROD/static/CollegeMedicalCenter/ein_CollegeMedicalCenter_standardcharges.csv.zip"
filename = url.split("/")[-1]
```

```python
df = pd.read_csv('46-2760935_CollegeMedicalCenter_standardcharges.csv', 
                 dtype={'CPT_CODE': str, 'REV_CODE': str, 'DRG_CODE': str, 'MODIFIER': str},
                 skiprows=3)
df
```

```python
df = df.rename(columns={
    'PRIMARY_PROCEDURE_DESCRIPTION': 'description',
    'SETTING_TYPE': 'patient_class',
    'DRG_CODE': 'drg',
    'CPT_CODE': 'cpt',
    'PAYER_NAME': 'payer_name',
    'NETWORK_NAME': 'plan_name',
    'MODIFIER': 'modifier',
    'REV_CODE': 'rev_code',
})

df['patient_class'] = df['patient_class'].str.lower()
df
```

```python
del df['SERVICE_COMPONENT_BREAKOUT_TYPE']
del df['SERVICE_COMPONENT_BREAKOUT_NAME']
del df['HCPCS_CODE']
df
```

```python
money_columns = df.columns.to_list()[6:-2]
remaining_columns = list(set(df.columns.to_list()) - set(money_columns))
df = pd.melt(df, id_vars=remaining_columns, var_name='payer_orig', value_name='rate')
df
```

```python
def payer_category_from_payer_orig(payer_orig):
    if payer_orig == 'GROSS_PRICE':
        return 'gross'
    elif payer_orig == 'CASH_PRICE':
        return 'cash'
    elif payer_orig == 'NEGOTIATED_PRICE':
        return 'payer'
    elif payer_orig == 'DEIDENTIFIED_MAX_PRICE':
        return 'max'
    elif payer_orig == 'DEIDENTIFIED_MIN_PRICE':
        return 'min'

df['payer_category'] = df['payer_orig'].apply(payer_category_from_payer_orig)
```

```python
df['code'] = 'na'
df['code_prefix'] = 'none'
df.loc[df['drg'].notnull(), 'code'] = df[df['drg'].notnull()]['drg']
df.loc[df['drg'].notnull(), 'code_prefix'] = 'ms-drg'
df.loc[df['cpt'].notnull(), 'code'] = df[df['cpt'].notnull()]['cpt']
df.loc[df['cpt'].notnull(), 'code_prefix'] = 'hcpcs_cpt'
df.loc[df['code_prefix'] == 'ms-drg', 'code_type'] = 'drg'
df.loc[df['code_prefix'] == 'hcpcs_cpt', 'code_type'] = 'cpt'
df
```

```python
del df['cpt']
del df['drg']
```

```python
df['hospital_ccn'] = ccn
df['filename'] = filename
df['url'] = url
df['hospital_ein'] = ein
df['internal_code'] = 'na'
df['modifier'] = df['modifier'].fillna('na')
df['ndc'] = 'na'
df['rev_code'] = df['rev_code'].fillna('na')
df['billing_class'] = 'na'
df['plan_orig'] = 'na'
df['patient_class'] = df['patient_class'].fillna('na')
df.loc[df['patient_class'] != 'payer', 'plan_name'] = None
df.loc[df['payer_orig'] == 'NEGOTIATED_PRICE', 'payer_orig'] = df[df['payer_orig'] == 'NEGOTIATED_PRICE']['payer_name']
df
```

```python
df_out = pd.DataFrame(columns=TARGET_COLUMNS)
df_out = df_out.append(df)
df_out = df_out.dropna(subset=['rate'], axis=0)
df_out.drop_duplicates()
df_out
```

```python
df_out.to_csv(ccn + ".csv", index=False)
```
