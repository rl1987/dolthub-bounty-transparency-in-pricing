```python
import pandas as pd
```

```python
TARGET_COLUMNS = [ 
    'hospital_ccn',
    'hospital_ein',
    'code_type',
    'internal_code',
    'description',
    'code_orig',
    'code_prefix',
    'code',
    'modifier',
    'ndc',
    'ndc_num_hcpcs_doses',
    'ndc_quantity_desc',
    'apc',
    'rev_code',
    'rev_desc',
    'billing_class',
    'patient_class',
    'payer_category',
    'payer_orig',
    'payer_name',
    'plan_orig',
    'plan_name',
    'plan_id',
    'plan_type',
    'rate',
    'rate_method',
    'rate_desc',
    'filename',
    'file_last_updated',
    'url',
    'permalink'
]
```

```python
ccn = "104024"
ein = "591101553"
filename = "Hospital-Standard-Charges.csv"
url = "http://05cdc38.rcomhost.com/CoC/wp-content/uploads/2021/01/Hospital-Standard-Charges.csv"
```

```python
df_in = pd.read_csv(filename, header=3)
df_in
```

```python
df_mid = pd.DataFrame(df_in)
df_mid = df_mid.rename(columns={
    'Payer': 'payer_orig',
    'Service': 'description',
    'RevCode': 'rev_code',
    'NDC': 'ndc',
    #'Gross Charge': 'gross',
    #'Negotiated Rate': 'rate',
    #'Discounted Cash Rate': 'cash',
    #'De-Identified Lowest': 'min',
    #'De-Identified Highest': 'max'
})
df_mid
```

```python
columns = df_mid.columns.to_list()
money_columns = ['Gross Charge', 'Discounted Cash Rate', 'De-Identified Lowest', 'De-Identified Highest', 'Negotiated Rate']
remaining_cols = list(set(columns) - set(money_columns))
df_mid = pd.melt(df_mid, id_vars=remaining_cols)
df_mid['variable'] = df_mid['variable'].replace('rate', 'payer')
df_mid = df_mid.rename(columns={
    'variable': 'payer_category',
    'value': 'rate',
})
df_mid['payer_orig'] = df_mid.apply(lambda row: row['payer_orig'] if row['payer_category'] == 'Negotiated Rate' else row['payer_category'], 
                                    axis=1)
df_mid['payer_category'] = df_mid['payer_category'].replace(
    'Gross Charge', 'gross').replace('Discounted Cash Rate', 'cash').replace(
    'De-Identified Lowest', 'min').replace('De-Identified Highest').replace(
    'Negotiated Rate', 'payer')
df_mid['rate'] = df_mid['rate'].astype(str)
df_mid['rate'] = df_mid['rate'].apply(lambda rate: rate.replace("$", "").replace(",", ""))
df_mid['ndc'] = 'na'
df_mid['rev_code'] = df_mid['rev_code'].astype(str)
df_mid['rev_code'] = df_mid['rev_code'].apply(lambda rev_code: "0" + rev_code)
df_mid
```

```python
df_mid['plan_orig'] = 'na'
df_mid['billing_class'] = 'na'
df_mid['patient_class'] = 'na'
df_mid['internal_code'] = 'na'
df_mid['code_prefix'] = 'none'
df_mid['code'] = df_mid['rev_code']
df_mid['modifier'] = 'na'
df_mid
```

```python
df_mid[df_mid['payer_category'] == 'cash']
```

```python
df_mid = df_mid.dropna(subset=['rate'])
df_mid = df_mid[df_mid['rate'] != 'nan']
df_mid
```

```python
in_f = open(filename, "r")
date_line = in_f.read().split("\n")[1]
in_f.close()
date_line
```

```python
updated_at = date_line.split("Effective ")[-1].split(",")[0]
day, month, year = updated_at.split("/")
updated_at = "{}-{}-{}".format(year, month, day)
df_mid['file_last_updated'] = updated_at
df_mid['hospital_ccn'] = ccn
df_mid['hospital_ein'] = ein[:2] + "-" + ein[2:]
df_mid['filename'] = filename
df_mid['url'] = url
df_mid
```

```python
df_out = pd.DataFrame(columns=TARGET_COLUMNS)
df_out = df_out.append(df_mid)
df_out
```

```python
df_out.to_csv('104024.csv', index=False)
```
