```python
import pandas as pd
```

```python
TARGET_COLUMNS = [
    "hospital_id",
    #'row_id',
    "line_type",
    "description",
    "rev_code",
    "local_code",
    "code",
    "ms_drg",
    "apr_drg",
    "eapg",
    "hcpcs_cpt",
    "modifiers",
    "alt_hcpcs_cpt",
    "thru",
    "apc",
    "icd",
    "ndc",
    "drug_hcpcs_multiplier",
    "drug_quantity",
    "drug_unit_of_measurement",
    "drug_type_of_measurement",
    "billing_class",
    "setting",
    "rate_category",
    "payer_name",
    "plan_name",
    "standard_charge",
    "standard_charge_percent",
    "contracting_method",
    "additional_generic_notes",
    "additional_payer_specific_notes",
]
```

```python
ccn = "104024"
ein = "591101553"
filename = "Hospital-Standard-Charges.csv"
url = "http://05cdc38.rcomhost.com/CoC/wp-content/uploads/2021/01/Hospital-Standard-Charges.csv"
```

```python
import subprocess

subprocess.run(["wget", "--no-clobber", url, "-O", filename])
```

```python
df_in = pd.read_csv(filename, header=3)
df_in
```

```python
df_mid = pd.DataFrame(df_in)
df_mid = df_mid.rename(columns={
    'Payer': 'payer_name',
    'Service': 'description',
    'RevCode': 'rev_code',
    'NDC': 'ndc',
})
df_mid
```

```python
columns = df_mid.columns.to_list()
money_columns = columns[4:]
remaining_cols = list(set(columns) - set(money_columns))
df_mid = pd.melt(df_mid, id_vars=remaining_cols)
df_mid['variable'] = df_mid['variable'].replace('rate', 'payer')
df_mid = df_mid.rename(columns={
    'variable': 'rate_category',
    'value': 'standard_charge',
})
df_mid['payer_name'] = df_mid.apply(lambda row: row['payer_name'] if row['rate_category'] == 'Negotiated Rate' else row['rate_category'], 
                                    axis=1)

df_mid
```

```python
set(df_mid['rate_category'].to_list())
```

```python
df_mid.loc[df_mid['rate_category'] == 'De-Identified Highest', 'rate_category'] = 'max'
df_mid.loc[df_mid['rate_category'] == 'De-Identified Lowest', 'rate_category'] = 'min'
df_mid.loc[df_mid['rate_category'] == 'Discounted Cash Rate', 'rate_category'] = 'cash'
df_mid.loc[df_mid['rate_category'] == 'Gross Charge', 'rate_category'] = 'gross'
df_mid.loc[df_mid['rate_category'] == 'Negotiated Rate', 'rate_category'] = 'negotiated'

df_mid['standard_charge'] = df_mid['standard_charge'].astype(str)
df_mid['standard_charge'] = df_mid['standard_charge'].apply(lambda rate: rate.replace("$", "").replace(",", ""))
df_mid['ndc'] = None
df_mid['rev_code'] = df_mid['rev_code'].astype(str)
df_mid['rev_code'] = df_mid['rev_code'].apply(lambda rev_code: "0" + rev_code)
df_mid
```

```python
df_mid['hospital_id'] = ccn

df_mid['line_type'] = None
df_mid['ms_drg'] = None
df_mid['apr_drg'] = None
df_mid['eapg'] = None
df_mid['hcpcs_cpt'] = None
df_mid['alt_hcpcs_cpt'] = None
df_mid['thru'] = None
df_mid['apc'] = None
df_mid['icd'] = None
df_mid['drug_hcpcs_multiplier'] = None
df_mid['drug_unit_of_measurement'] = None
df_mid['drug_type_of_measurement'] = None
df_mid['standard_charge_percent'] = None
df_mid['contracting_method'] = None
df_mid['additional_generic_notes'] = None
df_mid['additional_payer_specific_notes'] = None
df_mid['drug_quantity'] = None
df_mid['plan_name'] = None
df_mid['billing_class'] = None
df_mid['setting'] = None
df_mid['local_code'] = None
df_mid['code'] = df_mid['rev_code']
df_mid['modifiers'] = None

df_mid
```

```python
df_mid = df_mid.dropna(subset=['standard_charge'])
df_mid = df_mid[df_mid['standard_charge'] != 'nan']
df_mid
```

```python
in_f = open(filename, "r")
date_line = in_f.read().split("\n")[1]
in_f.close()
date_line
```

```python
df_out = pd.DataFrame(df_mid[TARGET_COLUMNS])
df_out
```

```python
df_out.to_csv('104024.csv', index=False)
```
