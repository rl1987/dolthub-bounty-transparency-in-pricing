```python
import json
import subprocess

import pandas as pd
```

```python
TARGET_COLUMNS = [ 
    'hospital_id',
    #'row_id',
    'line_type',
    'description',
    'rev_code',
    'local_code',
    'code',
    'ms_drg',
    'apr_drg',
    'eapg',
    'hcpcs_cpt',
    'modifiers',
    'alt_hcpcs_cpt',
    'thru',
    'apc',
    'icd',
    'ndc',
    'drug_hcpcs_multiplier',
    'drug_quantity',
    'drug_unit_of_measurement',
    'drug_type_of_measurement',
    'billing_class',
    'setting',
    'payer_category',
    'payer_name',
    'plan_name',
    'standard_charge',
    'standard_charge_percent',
    'contracting_method',
    'additional_generic_notes',
    'additional_payer_specific_notes'
]
```

```python
# https://www.sharp.com/patient/billing/pricing.cfm

#ccn = "050222"
#url = "https://images.sharp.com/com/patient/billing/952367304_sharp-chula-vista-medical-center_standardcharges.json"

#ccn = "050100"
#ccn = "050722"
ccn = "054145"
url = "https://images.sharp.com/com/patient/billing/953782169_sharp-memorial-hospital_standardcharges.json"

#ccn = "050234"
#url = "https://images.sharp.com/com/patient/billing/950651579_sharp-coronado-hospital-and-hlthcr-ctr_standardcharges.json"

#ccn = "050026"
#url = "https://images.sharp.com/com/patient/billing/330449527_grossmont-hospital_standardcharges.json"

filename = url.split("/")[-1].replace("%20", " ")
ein = filename.split("_")[0]
ein = ein[:2] + "-" + ein[2:]
ein
```

```python
subprocess.run(["wget", "--no-clobber", url, "-O", filename])
```

```python
in_f = open(filename, "r")
json_dict = json.load(in_f)
in_f.close()

list(json_dict.keys())
```

```python
df_in = pd.DataFrame(json_dict['Sheet1'])
df_in
```

```python
df_mid = pd.DataFrame(df_in)
df_mid = df_mid.rename(columns={
    'Charge\nCode': 'local_code',
    'Description': 'description',
    'MS\nDRG': 'ms_drg',
    'APR\nDRG': 'apr_drg',
    'Rev\nCode': 'rev_code',
    'CPT/\nHCPCS': 'hcpcs_cpt',
    'NDC': 'ndc',
    'Mod': 'modifiers'
})

df_mid.columns.to_list()
```

```python
remaining_columns = ['local_code', 'description', 'ms_drg', 'apr_drg', 'rev_code',
                     'hcpcs_cpt', 'ndc', 'modifiers']
df_mid = pd.melt(df_mid, id_vars=remaining_columns, var_name='payer_name', value_name='standard_charge')
df_mid
```

```python
df_mid['ms_drg'] = df_mid['ms_drg'].astype(str)
df_mid['ms_drg'] = df_mid['ms_drg'].apply(
    lambda ms_drg: ms_drg.replace('.0', '').zfill(3) if ms_drg != 'nan' else None
)
set(df_mid['ms_drg'].to_list())
```

```python
df_mid['rev_code'] = df_mid['rev_code'].astype(str)
df_mid['rev_code'] = df_mid['rev_code'].apply(
    lambda rev_code: rev_code.replace('.0', '').zfill(4) if rev_code != 'nan' else None
)
set(df_mid['rev_code'].to_list())
```

```python
def payer_category_from_payer_name(payer_name):
    if payer_name.startswith("Gross"):
        return 'gross'
    elif payer_name.startswith('Self Pay'):
        return 'cash'
    elif payer_name.startswith('Minimum'):
        return 'min'
    elif payer_name.startswith('Maximum'):
        return 'max'

    return 'payer'

df_mid['payer_category'] = df_mid['payer_name'].apply(payer_category_from_payer_name)
df_mid
```

```python
def setting_from_payer_name(payer_name):
    if type(payer_name) != str:
        return None

    if "Inpatient" in payer_name:
        return 'inpatient'
    elif "Outpatient" in payer_name:
        return 'outpatient'

    return None

df_mid['setting'] = df_mid['payer_name'].apply(setting_from_payer_name)
df_mid
```

```python
df_mid['local_code'] = df_mid['local_code'].astype(str)
df_mid.loc[~df_mid['local_code'].str.isnumeric(), 'line_type'] = df_mid[~df_mid['local_code'].str.isnumeric()]['local_code']
df_mid.loc[~df_mid['local_code'].str.isnumeric(), 'local_code'] = None
df_mid
```

```python
df_mid['hospital_id'] = ccn
df_mid['code'] = None
df_mid['eapg'] = None
df_mid['alt_hcpcs_cpt'] = None
df_mid['thru'] = None
df_mid['apc'] = None
df_mid['icd'] = None
df_mid['drug_hcpcs_multiplier'] = None
df_mid['drug_quantity'] = None
df_mid['drug_unit_of_measurement'] = None
df_mid['drug_type_of_measurement'] = None
df_mid['billing_class'] = None
df_mid['plan_name'] = None
df_mid['standard_charge_percent'] = None
df_mid['contracting_method'] = None
df_mid['additional_generic_notes'] = None
df_mid['additional_payer_specific_notes'] = None
```

```python
df_mid['apr_drg'] = df_mid['apr_drg'].astype(str)

df_mid['code'] = df_mid['apr_drg'].apply(
    lambda apr_drg: apr_drg if type(apr_drg) == str and len(apr_drg) == 5 and not '-' in apr_drg else None
)

df_mid['apr_drg'] = df_mid['apr_drg'].apply(
    lambda apr_drg: None if type(apr_drg) == str and len(apr_drg) == 5 and not '-' in apr_drg else apr_drg
)

df_mid.loc[df_mid['apr_drg'] == 'nan', 'apr_drg'] = None
```

```python
df_mid.loc[df_mid['standard_charge'] == 'N/A', 'standard_charge'] = None
df_mid.dropna(subset=['standard_charge'], inplace=True)
```

```python
df_out = pd.DataFrame(df_mid[TARGET_COLUMNS])
df_out
```

```python
df_out.to_csv('rate_' + ccn + '.csv', index=False)
```
