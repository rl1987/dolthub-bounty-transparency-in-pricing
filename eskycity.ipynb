```python
import pandas as pd
```

```python
TARGET_COLUMNS = [ 
    'hospital_id',
    #'row_id',
    'line_type',
    'description',
    'rev_code',
    'local_code',
    'code',
    'ms_drg',
    'apr_drg',
    'eapg',
    'hcpcs_cpt',
    'modifiers',
    'thru',
    'apc',
    'icd',
    'ndc',
    'drug_hcpcs_multiplier',
    'drug_quantity',
    'drug_unit_of_measurement',
    'drug_type_of_measurement',
    'billing_class',
    'setting',
    'payer_category',
    'payer_name',
    'plan_name',
    'standard_charge',
    'standard_charge_percent',
    'contracting_method',
    'additional_generic_notes',
    'additional_payer_specific_notes'
]
```

```python
ccn = "090001"
url = "https://uhsfilecdn.eskycity.net/ac/232896725_george-washington-university-hospital_standardcharges.csv"
filename = url.split("/")[-1]
ein = filename.split("_")[0]
ein = ein[:2] + "-" + ein[2:]
```

```python
import subprocess
subprocess.run(["wget", "--no-clobber", url, "-O", filename])
```

```python
df_in = pd.read_csv(filename, skiprows=3, dtype=str, low_memory=True)
df_in
```

```python
df_mid = pd.DataFrame(df_in)
df_mid = df_mid.rename(columns={
    'Description': 'description',
    'CDM': 'local_code',
    'CDM (If Applicable)': 'local_code',
    'Code Type': 'line_type',
    'DRG (If Applicable)': 'drg',
    'CPT/HCPCS (If Applicable)': 'hcpcs_cpt',
    'EAPG (If Applicable)': 'eapg',
    'APC (If Applicable)': 'apc',
    'Rev Code (If Applicable)': 'rev_code',
    'DRG (If applicable)': 'drg',
    'CPT/HCPCS (If applicable)': 'hcpcs_cpt',
    'EAPG (If applicable)': 'eapg',
    'APC (If applicable)': 'apc',
    'Rev Code (If applicable)': 'rev_code'
})
del df_mid['Facility']
df_mid = df_mid[df_mid.columns.drop(list(df_mid.filter(regex='Unnamed')))]
df_mid.columns = df_mid.columns.str.strip()
df_mid
```

```python
money_idx = df_mid.columns.to_list().index('Gross Charge')
money_columns = df_mid.columns.to_list()[money_idx:]
print(money_columns)
remaining_columns = df_mid.columns.to_list()[:money_idx]
df_mid = pd.melt(df_mid, id_vars=remaining_columns, var_name='payer_name', value_name='standard_charge')
df_mid
```

```python
df_mid.dropna(subset=['standard_charge'], inplace=True)
df_mid['standard_charge'] = df_mid['standard_charge'].astype(str)
df_mid['standard_charge'] = df_mid['standard_charge'].str.strip()
df_mid['standard_charge'] = df_mid['standard_charge'].apply(lambda rate: rate.replace("$", "").replace(",", "").strip())
df_mid = df_mid[~df_mid['standard_charge'].str.contains("\(1\)")]
df_mid = df_mid[df_mid['standard_charge'] != "-"]
df_mid = df_mid[df_mid['standard_charge'] != "v"]
df_mid = df_mid[df_mid['standard_charge'] != 'Not Reimb']
df_mid = pd.DataFrame(df_mid)
df_mid
```

```python
df_mid.loc[df_mid['standard_charge'].str.contains("  Per diem"), 'contracting_method'] = "per diem"
df_mid.loc[df_mid['standard_charge'].str.contains("  Per diem"), 'standard_charge'] = df_mid[df_mid['standard_charge'].str.contains("  Per diem")]['standard_charge'].str.replace(" Per diem", "")
df_mid['standard_charge'] = df_mid['standard_charge'].str.strip()

df_mid.loc[df_mid['standard_charge'].str.contains(" per diem"), 'contracting_method'] = "per diem"
df_mid.loc[df_mid['standard_charge'].str.contains(" per diem"), 'standard_charge'] = df_mid[df_mid['standard_charge'].str.contains(" per diem")]['standard_charge'].str.replace(" per diem", "")
df_mid.loc[df_mid['standard_charge'].str.contains(" Per Diem"), 'contracting_method'] = "per diem"
df_mid.loc[df_mid['standard_charge'].str.contains(" Per Diem"), 'standard_charge'] = df_mid[df_mid['standard_charge'].str.contains(" Per Diem")]['standard_charge'].str.replace(" Per Diem", "")
df_mid.loc[df_mid['standard_charge'].str.contains(" Per Deim"), 'contracting_method'] = "per diem"
df_mid.loc[df_mid['standard_charge'].str.contains(" Per Deim"), 'standard_charge'] = df_mid[df_mid['standard_charge'].str.contains(" Per Deim")]['standard_charge'].str.replace(" Per Deim", "")
df_mid.loc[df_mid['standard_charge'].str.contains(" Per Deiem"), 'contracting_method'] = "per diem"
df_mid.loc[df_mid['standard_charge'].str.contains(" Per Deiem"), 'standard_charge'] = df_mid[df_mid['standard_charge'].str.contains(" Per Deiem")]['standard_charge'].str.replace(" Per Deiem", "")

df_mid.loc[df_mid['standard_charge'].str.contains("% of Charges"), 'contracting_method'] = "percent of total billed charge"
df_mid.loc[df_mid['standard_charge'].str.contains("% of Charges"), 'standard_charge_percent'] = df_mid[df_mid['standard_charge'].str.contains("% of Charges")]['standard_charge'].str.replace("% of Charges", "")
df_mid.loc[df_mid['standard_charge'].str.contains("% of covered Charges"), 'contracting_method'] = "percent of total billed charge"
df_mid.loc[df_mid['standard_charge'].str.contains("% of covered Charges"), 'standard_charge_percent'] = df_mid[df_mid['standard_charge'].str.contains("% of covered Charges")]['standard_charge'].str.replace("% of covered Charges", "")
df_mid.loc[df_mid['standard_charge'].str.contains("Charges"), 'standard_charge'] = None

df_mid
```

```python
df_mid['drg'] = df_mid['drg'].fillna('na')

def pad_drg_if_needed(drg):
    if drg == 'na':
        return drg
    
    if len(drg) == 1:
        return '00' + drg
    elif len(drg) == 2:
        return '0' + drg
    
    return drg

df_mid['drg'] = df_mid['drg'].apply(pad_drg_if_needed)
df_mid.loc[df_mid['drg'] == 'na', 'drg'] = None
df_mid
```

```python
if 'line_type' in df_mid.columns:
    df_mid.loc[df_mid['line_type'] == 'APR-DRG', 'apr_drg'] = df_mid[df_mid['line_type'] == 'APR-DRG']['drg']
    df_mid.loc[df_mid['line_type'] == 'APR-DRG', 'code'] = df_mid[df_mid['line_type'] == 'APR-DRG']['drg']
    df_mid.loc[df_mid['line_type'] == 'MS-DRG', 'ms_drg'] = df_mid[df_mid['line_type'] == 'MS-DRG']['drg']
    df_mid.loc[df_mid['line_type'] == 'MS-DRG', 'code'] = df_mid[df_mid['line_type'] == 'MS-DRG']['drg']
    del df_mid['drg']
    df_mid.loc[df_mid['line_type'] == 'Chargemaster', 'code'] = df_mid[df_mid['line_type'] == 'Chargemaster']['local_code']
    df_mid.loc[df_mid['line_type'] == 'EAPG', 'code'] = df_mid[df_mid['line_type'] == 'EAPG']['eapg']
    df_mid.loc[df_mid['line_type'] == 'Other CPT/HCPCS', 'code'] = df_mid[df_mid['line_type'] == 'Other CPT/HCPCS']['hcpcs_cpt']
else:
    df_mid = df_mid.rename(columns={'drg': 'ms_drg'})

df_mid['thru'] = None
df_mid['icd'] = None
df_mid['ndc'] = None
df_mid['drug_hcpcs_multiplier'] = None
df_mid['drug_quantity'] = None
df_mid['drug_unit_of_measurement'] = None
df_mid['drug_type_of_measurement'] = None
df_mid['additional_generic_notes'] = None
df_mid['additional_payer_specific_notes'] = None
```

```python
if 'line_type' in df_mid.columns:
    set(df_mid['line_type'].tolist())
else:
    df_mid['line_type'] = None
    
if not 'code' in df_mid.columns:
    df_mid['code'] = None
    
if not 'apr_drg' in df_mid.columns:
    df_mid['apr_drg'] = None
    
if not 'eapg' in df_mid.columns:
    df_mid['eapg'] = None
```

```python
set(df_mid['apc'].to_list()) # TODO: make sure the APCs fit the format.
```

```python
df_mid['modifiers'] = None

df_mid['hcpcs_cpt'] = df_mid['hcpcs_cpt'].fillna('')
df_mid.loc[df_mid['hcpcs_cpt'].str.len() == 7, 'modifiers'] = df_mid[df_mid['hcpcs_cpt'].str.len() == 7]['hcpcs_cpt'].str[-2:]
df_mid.loc[df_mid['hcpcs_cpt'].str.len() == 7, 'hcpcs_cpt'] = df_mid[df_mid['hcpcs_cpt'].str.len() == 7]['hcpcs_cpt'].str[:5]
df_mid.loc[df_mid['hcpcs_cpt'] == '', 'hcpcs_cpt'] = None
df_mid.loc[df_mid['hcpcs_cpt'] == 'XXXXX', 'hcpcs_cpt'] = None
df_mid.loc[df_mid['hcpcs_cpt'] == 'C1C13', 'hcpcs_cpt'] = None

df_mid['modifiers'].value_counts()
```

```python
df_mid['payer_name'] = df_mid['payer_name'].apply(lambda payer_orig: payer_orig.strip())

def payer_category_from_payer_orig(payer_orig):
    if payer_orig == "Cash Price":
        return 'cash'
    elif payer_orig == "Gross Charge":
        return 'gross'
    elif payer_orig == "Minimum":
        return 'min'
    elif payer_orig == "Maximum":
        return 'max'
    
    return 'payer'

df_mid['payer_category'] = df_mid['payer_name'].apply(payer_category_from_payer_orig)
df_mid
```

```python
from helpers import pad_rev_code_if_needed
if 'rev_code' in df_mid.columns:
    df_mid['rev_code'] = df_mid['rev_code'].apply(pad_rev_code_if_needed)
else:
    df_mid['rev_code'] = None
df_mid['billing_class'] = None
df_mid['setting'] = None
df_mid['plan_name'] = None
df_mid
```

```python
df_mid['hospital_id'] = ccn
```

```python
df_out = pd.DataFrame(df_mid[TARGET_COLUMNS])
df_out
```

```python
df_out.to_csv("eskycity_" + ccn + ".csv", index=False)
```
