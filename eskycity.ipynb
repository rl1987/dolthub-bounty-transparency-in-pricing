```python
import pandas as pd
```

```python
TARGET_COLUMNS = [ 
    'hospital_ccn',
    'hospital_ein',
    'code_type',
    'internal_code',
    'description',
    'code_orig',
    'code_prefix',
    'code',
    'modifier',
    'ndc',
    'ndc_num_hcpcs_doses',
    'ndc_quantity_desc',
    'apc',
    'rev_code',
    'rev_desc',
    'billing_class',
    'patient_class',
    'payer_category',
    'payer_orig',
    'payer_name',
    'plan_orig',
    'plan_name',
    'plan_id',
    'plan_type',
    'rate',
    'rate_method',
    'rate_desc',
    'filename',
    'file_last_updated',
    'url',
    'permalink'
]
```

```python
ccn = "450119"
url = "https://uhsfilecdn.eskycity.net/ac/233069260_south-texas-health-system-mcallen_standardcharges.csv"
filename = url.split("/")[-1]
ein = filename.split("_")[0]
ein = ein[:2] + "-" + ein[2:]
```

```python
import subprocess
subprocess.run(["wget", "--no-clobber", url, "-O", filename])
```

```python
df_in = pd.read_csv(filename, skiprows=3, dtype=str, low_memory=True)
df_in
```

```python
df_mid = pd.DataFrame(df_in)
df_mid = df_mid.rename(columns={
    'Description': 'description',
    'CDM': 'cdm',
    'Code Type': 'code_type',
    'DRG (If Applicable)': 'drg',
    'CPT/HCPCS (If Applicable)': 'cpt',
    'EAPG (If Applicable)': 'eapg',
    'APC (If Applicable)': 'apc',
    'Rev Code (If Applicable)': 'rev_code'
})
del df_mid['Facility']
df_mid
```

```python
money_columns = df_mid.columns.to_list()[8:]
remaining_columns = df_mid.columns.to_list()[:8]
df_mid = pd.melt(df_mid, id_vars=remaining_columns, var_name='payer_orig', value_name='rate')
df_mid
```

```python
df_mid['rate'] = df_mid['rate'].str.strip()
df_mid['rate'] = df_mid['rate'].apply(lambda rate: rate.replace("$", "").replace(",", ""))
df_mid = df_mid[~df_mid['rate'].str.contains("\(1\)")]
df_mid = df_mid[df_mid['rate'] != "-"]
df_mid = pd.DataFrame(df_mid)
df_mid
```

```python
set(df_mid['code_type'].tolist())
```

```python
df_mid['drg'] = df_mid['drg'].fillna('na')

def pad_drg_if_needed(drg):
    if drg == 'na':
        return drg
    
    if len(drg) == 1:
        return '00' + drg
    elif len(drg) == 2:
        return '0' + drg
    
    return drg

df_mid['drg'] = df_mid['drg'].apply(pad_drg_if_needed)
```

```python
df_mid['code_prefix'] = 'none'
df_mid['code_prefix'] = df_mid['code_type'].replace('APR-DRG', 'apr-drg').replace(
    'EAPG', 'eapg').replace('MS-DRG', 'ms-drg').replace('Other CPT/HCPCS', 'hcpcs_cpt').replace(
    'Chargemaster', 'cdm')
df_mid['code_type'] = df_mid['code_type'].replace('APR-DRG', 'drg').replace(
    'EAPG', 'eap').replace('MS-DRG', 'drg').replace('Other CPT/HCPCS', 'cpt').replace(
    'Chargemaster', 'cdm')
df_mid
```

```python
df_mid.loc[df_mid['code_type'] == 'drg', 'code'] = df_mid[df_mid['code_type'] == 'drg']['drg']
del df_mid['drg']
df_mid.loc[df_mid['code_type'] == 'eap', 'code'] = df_mid[df_mid['code_type'] == 'eap']['eapg']
del df_mid['eapg']
df_mid.loc[df_mid['code_type'] == 'cpt', 'code'] = df_mid[df_mid['code_type'] == 'cpt']['cpt']
del df_mid['cpt']
df_mid.loc[df_mid['code_type'] == 'cdm', 'code'] = df_mid[df_mid['code_type'] == 'cdm']['cdm']
del df_mid['cdm']
df_mid.loc[df_mid['code_type'] == 'apc', 'code'] = df_mid[df_mid['code_type'] == 'apc']['apc']
```

```python
df_mid
```

```python
set(df_mid['payer_orig'].tolist())
```

```python
df_mid['payer_orig'] = df_mid['payer_orig'].apply(lambda payer_orig: payer_orig.strip())

def payer_category_from_payer_orig(payer_orig):
    if payer_orig == "Cash Price":
        return 'cash'
    elif payer_orig == "Gross Charge":
        return 'gross'
    elif payer_orig == "Minimum":
        return 'min'
    elif payer_orig == "Maximum":
        return 'max'
    
    return 'payer'

df_mid['payer_category'] = df_mid['payer_orig'].apply(payer_category_from_payer_orig)
df_mid
```

```python
df_mid['internal_code'] = 'na'
df_mid['modifier'] = 'na'
df_mid['ndc'] = 'na'
df_mid['rev_code'] = df_mid['rev_code'].fillna('na')
from helpers import pad_rev_code_if_needed
df_mid['rev_code'] = df_mid['rev_code'].apply(pad_rev_code_if_needed)
df_mid['billing_class'] = 'na'
df_mid['patient_class'] = 'na'
df_mid['plan_orig'] = 'na'
df_mid
```

```python
df_mid = df_mid.dropna(subset=['rate'], axis=0)

def rate_method_from_rate(rate):
    if "Per Diem" in rate:
        return "Per Diem"
    elif "%" in rate:
        return rate
    
    return None

df_mid['rate_method'] = df_mid['rate'].apply(rate_method_from_rate)
df_mid.loc[df_mid['rate_method'] == "Per Diem", 'rate'] = df_mid[df_mid['rate_method'] == "Per Diem"]['rate'].apply(lambda rate: rate.split(" ")[0])
df_mid['rate'] = df_mid.apply(lambda row: -1 if row['rate_method'] is not None and "%" in row['rate_method'] else row['rate'], axis=1)
df_mid
```

```python
df_mid[df_mid['rate_method'].notnull()].sample(20)
```

```python
df_mid['hospital_ccn'] = ccn
df_mid['hospital_ein'] = ein
df_mid['filename'] = filename
df_mid['url'] = url
df_mid
```

```python
df_out = pd.DataFrame(columns=TARGET_COLUMNS)
df_out = df_out.append(df_mid)
df_out
```

```python
df_out.to_csv(ccn + ".csv", index=False)
```
