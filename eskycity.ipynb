```python
import pandas as pd
```

```python
TARGET_COLUMNS = [ 
    'hospital_id',
    #'row_id',
    'line_type',
    'description',
    'rev_code',
    'local_code',
    'code',
    'ms_drg',
    'apr_drg',
    'eapg',
    'hcpcs_cpt',
    'modifiers',
    'thru',
    'apc',
    'icd',
    'ndc',
    'drug_hcpcs_multiplier',
    'drug_quantity',
    'drug_unit_of_measurement',
    'drug_type_of_measurement',
    'billing_class',
    'setting',
    'payer_category',
    'payer_name',
    'plan_name',
    'standard_charge',
    'standard_charge_percent',
    'contracting_method',
    'additional_generic_notes',
    'additional_payer_specific_notes'
]
```

```python
ccn = "100299"
url = "https://uhsfilecdn.eskycity.net/ac/731666433_lakewood-ranch-medical-center_standardcharges.csv"
filename = url.split("/")[-1]
ein = filename.split("_")[0]
ein = ein[:2] + "-" + ein[2:]
```

```python
import subprocess
subprocess.run(["wget", "--no-clobber", url, "-O", filename])
```

```python
df_in = pd.read_csv(filename, skiprows=3, dtype=str, low_memory=True)
df_in
```

```python
df_mid = pd.DataFrame(df_in)
df_mid = df_mid.rename(columns={
    'Description': 'description',
    'CDM': 'local_code',
    'Code Type': 'line_type',
    'DRG (If Applicable)': 'drg',
    'CPT/HCPCS (If Applicable)': 'hcpcs_cpt',
    'EAPG (If Applicable)': 'eapg',
    'APC (If Applicable)': 'apc',
    'Rev Code (If Applicable)': 'rev_code'
})
del df_mid['Facility']
df_mid
```

```python
money_columns = df_mid.columns.to_list()[8:]
remaining_columns = df_mid.columns.to_list()[:8]
df_mid = pd.melt(df_mid, id_vars=remaining_columns, var_name='payer_name', value_name='standard_charge')
df_mid
```

```python
df_mid['standard_charge'] = df_mid['standard_charge'].str.strip()
df_mid['standard_charge'] = df_mid['standard_charge'].apply(lambda rate: rate.replace("$", "").replace(",", ""))
df_mid = df_mid[~df_mid['standard_charge'].str.contains("\(1\)")]
df_mid = df_mid[df_mid['standard_charge'] != "-"]
df_mid = pd.DataFrame(df_mid)
df_mid
```

```python
set(df_mid['line_type'].tolist())
```

```python
df_mid['drg'] = df_mid['drg'].fillna('na')

def pad_drg_if_needed(drg):
    if drg == 'na':
        return drg
    
    if len(drg) == 1:
        return '00' + drg
    elif len(drg) == 2:
        return '0' + drg
    
    return drg

df_mid['drg'] = df_mid['drg'].apply(pad_drg_if_needed)
df_mid.loc[df_mid['drg'] == 'na', 'drg'] = None
```

```python
df_mid.loc[df_mid['line_type'] == 'APR-DRG', 'apr-drg'] = df_mid[df_mid['line_type'] == 'APR-DRG']['drg']
df_mid.loc[df_mid['line_type'] == 'APR-DRG', 'code'] = df_mid[df_mid['line_type'] == 'APR-DRG']['drg']
df_mid.loc[df_mid['line_type'] == 'MS-DRG', 'ms-drg'] = df_mid[df_mid['line_type'] == 'MS-DRG']['drg']
df_mid.loc[df_mid['line_type'] == 'MS-DRG', 'code'] = df_mid[df_mid['line_type'] == 'MS-DRG']['drg']
del df_mid['drg']
df_mid.loc[df_mid['line_type'] == 'Chargemaster', 'code'] = df_mid[df_mid['line_type'] == 'Chargemaster']['local_code']
df_mid.loc[df_mid['line_type'] == 'EAPG', 'code'] = df_mid[df_mid['line_type'] == 'EAPG']['eapg']
df_mid.loc[df_mid['line_type'] == 'Other CPT/HCPCS', 'code'] = df_mid[df_mid['line_type'] == 'Other CPT/HCPCS']['hcpcs_cpt']
df_mid
```

```python
df_mid['payer_name'] = df_mid['payer_name'].apply(lambda payer_orig: payer_orig.strip())

def payer_category_from_payer_orig(payer_orig):
    if payer_orig == "Cash Price":
        return 'cash'
    elif payer_orig == "Gross Charge":
        return 'gross'
    elif payer_orig == "Minimum":
        return 'min'
    elif payer_orig == "Maximum":
        return 'max'
    
    return 'payer'

df_mid['payer_category'] = df_mid['payer_name'].apply(payer_category_from_payer_orig)
df_mid
```

```python
df_mid['modifier'] = None
df_mid['ndc'] = None

from helpers import pad_rev_code_if_needed
df_mid['rev_code'] = df_mid['rev_code'].apply(pad_rev_code_if_needed)
df_mid['billing_class'] = None
df_mid['patient_class'] = None
df_mid['plan_name'] = None
df_mid
```

```python
df_mid = df_mid.dropna(subset=['rate'], axis=0)

def rate_method_from_rate(rate):
    if "Per Diem" in rate:
        return "Per Diem"
    elif "%" in rate:
        return rate
    
    return None

df_mid['rate_method'] = df_mid['rate'].apply(rate_method_from_rate)
df_mid.loc[df_mid['rate_method'] == "Per Diem", 'rate'] = df_mid[df_mid['rate_method'] == "Per Diem"]['rate'].apply(lambda rate: rate.split(" ")[0])
df_mid['rate'] = df_mid.apply(lambda row: -1 if row['rate_method'] is not None and "%" in row['rate_method'] else row['rate'], axis=1)
df_mid
```

```python
df_mid[df_mid['rate_method'].notnull()].sample(20)
```

```python
df_mid['hospital_ccn'] = ccn
df_mid['hospital_ein'] = ein
df_mid['filename'] = filename
df_mid['url'] = url
df_mid
```

```python
df_out = pd.DataFrame(columns=TARGET_COLUMNS)
df_out = df_out.append(df_mid)
df_out
```

```python
df_out.to_csv(ccn + ".csv", index=False)
```
