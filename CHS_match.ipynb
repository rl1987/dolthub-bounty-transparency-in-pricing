```python
import doltcli as dolt
import pandas as pd
```

```python
df = pd.read_csv('chs_urls.csv')
df
```

```python
df['phone'].value_counts()
```

```python
df['phone'] = df['phone'].apply(lambda phone: phone.replace(")", "").replace("(", "").replace(" ", "").replace("-", ""))
set(df['phone'].to_list())
```

```python
db = dolt.Dolt("/root/data/transparency-in-pricing")

def lookup_ccn_by_phone(phone):
    sql = "SELECT id FROM hospital WHERE phone = '{}';".format(phone)
    
    res = db.sql(sql, result_format="json")
    if res.get('rows') is None or len(res['rows']) != 1:
        return None

    return res['rows'][0]['id']

df['ccn'] = df['phone'].apply(lookup_ccn_by_phone)
df
```

```python
df[df['ccn'].isnull()]
```

```python
def lookup_ccn_by_name_city_state(name, city, state):
    sql = "SELECT id FROM hospital WHERE name LIKE '%{}%' AND city = '{}' AND state = '{}';".format(name, city, state)
    
    res = db.sql(sql, result_format="json")
    if res.get('rows') is None or len(res['rows']) != 1:
        return None

    return res['rows'][0]['id']

df.loc[df['ccn'].isnull(), 'ccn'] = df[df['ccn'].isnull()][['name', 'city', 'state']].apply(
    lambda row: lookup_ccn_by_name_city_state(row['name'], row['city'], row['state']),
    axis=1
)
```
