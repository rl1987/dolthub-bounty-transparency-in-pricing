```python
import pandas as pd
```

```python
df = pd.read_csv('http_head_results.csv')
df
```

```python
df = df[df['http_status'] == 200]
df = df.dropna(subset=['content_len'], axis=0)
df['content_len'] = df['content_len'].astype(int)
df = df[df['content_len'] < 100 * 1024 * 1024]
df = df[df['standard_charge_file_url'].str.contains("standardcharge")]
df
```
