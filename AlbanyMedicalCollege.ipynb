```python
import subprocess

import pandas as pd
```

```python
TARGET_COLUMNS = [ 
    'hospital_id',
    #'row_id',
    'line_type',
    'description',
    'rev_code',
    'local_code',
    'code',
    'ms_drg',
    'apr_drg',
    'eapg',
    'hcpcs_cpt',
    'modifiers',
    'alt_hcpcs_cpt',
    'thru',
    'apc',
    'icd',
    'ndc',
    'drug_hcpcs_multiplier',
    'drug_quantity',
    'drug_unit_of_measurement',
    'drug_type_of_measurement',
    'billing_class',
    'setting',
    'payer_category',
    'payer_name',
    'plan_name',
    'standard_charge',
    'standard_charge_percent',
    'contracting_method',
    'additional_generic_notes',
    'additional_payer_specific_notes'
]
```

```python
# https://www.albanymed.org/patients-visitors/billing-insurance/hospital-pricing-transparency/
ccn = "330013"
url = "https://www.albanymed.org/wp-content/uploads/attachements/141338307_AlbanyMedicalCenter_standardcharges.csv"
filename = url.split("/")[-1].replace("%20", " ")
ein = filename.split("_")[0]
ein = ein[:2] + "-" + ein[2:]
ein
```

```python
import subprocess
subprocess.run(["wget", "--no-clobber", url, "-O", filename])
```

```python
df_in = pd.read_csv(filename, encoding='latin-1', skiprows=[0,1], low_memory=False)
df_in
```

```python
delim_idx = df_in[df_in['Charge Code'] == 'Table 2: Payer Negotiated Rates'].index.to_list()[0]
delim_idx
```

```python
df_in1 = pd.DataFrame(df_in.head(delim_idx))
df_in1
```

```python
set(df_in1['Additonal Comment'].to_list())
```

```python
df_mid1 = pd.DataFrame(df_in1)
df_mid1 = df_mid1.rename(columns={
    'Charge Code': 'local_code',
    'Description': 'description',
    'CPT® Code': 'hcpcs_cpt',
    'Revenue Code': 'rev_code',
    'Additonal Comment': 'additional_generic_notes'
})

# https://stackoverflow.com/a/67894930
column_to_move = df_mid1.pop("additional_generic_notes")
df_mid1.insert(0, "additional_generic_notes", column_to_move)

df_mid1
```

```python
from helpers import cleanup_values

money_columns = df_mid1.columns[5:]
df_mid1[money_columns] = df_mid1[money_columns].apply(cleanup_values)
df_mid1
```

```python
df_mid1['description'] = df_mid1['description'].str.strip()
```

```python
remaining_columns = df_mid1.columns[:5]
df_mid1 = pd.melt(df_mid1, id_vars=remaining_columns, var_name='payer_name', value_name='standard_charge')
df_mid1
```

```python
def payer_category_from_payer_name(payer_name):
    if payer_name == "Gross Charge/Unit":
        return "gross"
    elif payer_name == "Cash Price":
        return "cash"
    elif payer_name == "MinRate" or payer_name == "Deidentified Min Rate":
        return "min"
    elif payer_name == "MaxRate" or payer_name == "Deidentified Max Rate":
        return "max"

    return "payer"

df_mid1['payer_category'] = df_mid1['payer_name'].apply(payer_category_from_payer_name)
df_mid1[['payer_name', 'payer_category']].value_counts()
```

```python
df_mid1.loc[df_mid1['payer_name'].str.endswith(' IP'), 'setting'] = 'inpatient'
df_mid1.loc[df_mid1['payer_name'].str.endswith(' OP'), 'setting'] = 'outpatient'
df_mid1[['payer_name', 'setting']].value_counts()
```

```python
df_mid1['hospital_id'] = ccn
df_mid1['line_type'] = None
df_mid1['code'] = None
df_mid1['ms_drg'] = None
df_mid1['apr_drg'] = None
df_mid1['eapg'] = None
df_mid1['modifiers'] = None
df_mid1['alt_hcpcs_cpt'] = None
df_mid1['thru'] = None
df_mid1['apc'] = None
df_mid1['icd'] = None
df_mid1['ndc'] = None
df_mid1['drug_hcpcs_multiplier'] = None
df_mid1['drug_quantity'] = None
df_mid1['drug_unit_of_measurement'] = None
df_mid1['drug_type_of_measurement'] = None
df_mid1['billing_class'] = None
df_mid1['plan_name'] = None
df_mid1['standard_charge_percent'] = None
df_mid1['contracting_method'] = None
df_mid1['additional_payer_specific_notes'] = None

df_mid1
```

```python
df_mid1 = df_mid1.dropna(subset=['standard_charge'], axis=0)

df_out1 = pd.DataFrame(df_mid1[TARGET_COLUMNS])
df_out1
```

```python
df_out1.to_csv('rate_' + ccn + '_tbl1.csv', index=False)
```

```python
df_in2 = pd.DataFrame(df_in.iloc[delim_idx+1:])
df_in2.columns = df_in2.iloc[0]
df_in2 = df_in2.iloc[1:]
df_in2.drop(columns=df_in2.columns[-5:], axis=1, inplace=True)
df_in2
```

```python
set(df_in2['Service'].to_list())
```

```python
set(df_in2['ReimburseType'].to_list())
```

```python
df_mid2 = pd.DataFrame(df_in2)
df_mid2 = df_mid2.rename(columns={
    'Payer Name': 'payer_name',
    'CodeType': 'line_type',
    'Code': 'code',
    'Description': 'description',
    'Service': 'additional_payer_specific_notes',
    'ReimburseType': 'contracting_method',
    'ServiceType': 'setting'
})

# https://stackoverflow.com/a/67894930
column_to_move = df_mid2.pop("setting")
df_mid2.insert(0, "setting", column_to_move)

df_mid2
```

```python
money_columns = df_mid2.columns[-3:]
remaining_columns = df_mid2.columns[:-3]
df_mid2 = pd.melt(df_mid2, id_vars=remaining_columns, var_name='payer_name2', value_name='standard_charge')
df_mid2
```

```python
df_mid2.loc[df_mid2['setting'] == 'IP', 'setting'] = 'inpatient'
df_mid2.loc[df_mid2['setting'] == 'OP', 'setting'] = 'outpatient'
df_mid2
```

```python
df_mid2['description'] = df_mid2['description'].str.strip()
```

```python
set(df_mid2['line_type'].to_list())
```

```python
df_mid2.loc[df_mid2['line_type'] == 'CPT', 'hcpcs_cpt'] = df_mid2[df_mid2['line_type'] == 'CPT']['code']
df_mid2['hcpcs_cpt'] = df_mid2['hcpcs_cpt'].fillna('')

# 0075T w Rev 320 or 321 or 322 or 323
df_mid2.loc[df_mid2['hcpcs_cpt'].str.contains(' w Rev '), 'rev_code'] = df_mid2[df_mid2['hcpcs_cpt'].str.contains(' w Rev ')]['hcpcs_cpt'].apply(
    lambda cpt: cpt.split(' w Rev ')[1].split(' or ')[0].zfill(4)
)
df_mid2.loc[df_mid2['hcpcs_cpt'].str.contains(' w Rev '), 'hcpcs_cpt'] = df_mid2[df_mid2['hcpcs_cpt'].str.contains(' w Rev ')]['hcpcs_cpt'].apply(
    lambda cpt: cpt.split(' w Rev ')[0]
)

df_mid2['hcpcs_cpt'] = df_mid2['hcpcs_cpt'].str.upper()

df_mid2.loc[df_mid2['hcpcs_cpt'].str.len() == 4, 'rev_code'] = df_mid2[df_mid2['hcpcs_cpt'].str.len() == 4]['hcpcs_cpt']
df_mid2.loc[df_mid2['hcpcs_cpt'].str.len() == 4, 'hcpcs_cpt'] = None
df_mid2.loc[df_mid2['hcpcs_cpt'] == '', 'hcpcs_cpt'] = None
```

```python
df_mid2.loc[df_mid2['line_type'] == 'DRG', 'ms_drg'] = df_mid2[df_mid2['line_type'] == 'DRG']['code'].str.replace('MS-', '')
df_mid2.loc[df_mid2['ms_drg'].str.len() == 4, 'apr_drg'] = df_mid2[df_mid2['ms_drg'].str.len() == 4]['ms_drg'].apply(
    lambda ms_drg: ms_drg[:3] + '-' + ms_drg[-1]
)
df_mid2.loc[df_mid2['ms_drg'].str.len() == 4, 'ms_drg'] = None
```

```python
df_mid2.loc[df_mid2['line_type'] == 'ICD Code', 'icd'] = df_mid2[df_mid2['line_type'] == 'ICD Code']['code']
```

```python
df_mid2.loc[df_mid2['line_type'] == 'Revenue Code', 'rev_code'] = df_mid2[df_mid2['line_type'] == 'Revenue Code']['code']

# 0481 - 92920
# Blue Shield Indemnity PPO,Revenue Code,0481 - 92920,Cardiology: q,Cardiac/Vascular Anglo,Per Case,25821.00,9710.00,25821.00,OP,,,,,
df_mid2['rev_code'] = df_mid2['rev_code'].fillna('')
df_mid2.loc[df_mid2['rev_code'].str.contains(' - '), 'hcpcs_cpt'] = df_mid2[df_mid2['rev_code'].str.contains(' - ')]['rev_code'].apply(
    lambda rev_code: rev_code.split(' - ')[-1]
)
df_mid2.loc[df_mid2['rev_code'].str.contains(' - '), 'rev_code'] = df_mid2[df_mid2['rev_code'].str.contains(' - ')]['rev_code'].apply(
    lambda rev_code: rev_code.split(' - ')[0]
)

# 636-J9280 w DRG 327,330,,331,357 or 358
# CDPHP Commercial,Revenue Code,"636-J9280 w DRG 327,330,,331,357 or 358",Mytomycin,HIPEC Add-On,Per Unit,6695.00,6695.00,6695.00,IP,,,,,
df_mid2.loc[df_mid2['rev_code'].str.contains(' w DRG '), 'hcpcs_cpt'] = df_mid2[df_mid2['rev_code'].str.contains(' w DRG ')]['rev_code'].apply(
    lambda rev_code: rev_code.split(' w DRG ')[0].split('-')[-1]
)
df_mid2.loc[df_mid2['rev_code'].str.contains(' w DRG '), 'ms_drg'] = df_mid2[df_mid2['rev_code'].str.contains(' w DRG ')]['rev_code'].apply(
    lambda rev_code: rev_code.split(' w DRG ')[-1].split(',')[0]
)
df_mid2.loc[df_mid2['rev_code'].str.contains(' w DRG '), 'rev_code'] = df_mid2[df_mid2['rev_code'].str.contains(' w DRG ')]['rev_code'].apply(
    lambda rev_code: rev_code.split(' w DRG ')[0].split('-')[0].zfill(4)
)

df_mid2.loc[df_mid2['rev_code'].str.len() == 5, 'hcpcs_cpt'] = df_mid2[df_mid2['rev_code'].str.len() == 5]['rev_code']
df_mid2.loc[df_mid2['rev_code'].str.len() == 5, 'rev_code'] = None
df_mid2
```

```python
df_mid2.loc[df_mid2['payer_name2'] != "Rate", 'payer_name'] = df_mid2[df_mid2['payer_name2'] != "Rate"]['payer_name2']
del df_mid2['payer_name2']

df_mid2
```

```python
df_mid2['payer_category'] = df_mid2['payer_name'].apply(payer_category_from_payer_name)
df_mid2
```

```python
df_mid2['additional_payer_specific_notes'] = df_mid2['additional_payer_specific_notes'].fillna('')
df_mid2.loc[df_mid2['additional_payer_specific_notes'].str.startswith('APC '), 'apc'] = df_mid2[df_mid2['additional_payer_specific_notes'].str.startswith('APC ')]['additional_payer_specific_notes'].str.replace('APC ', '')
df_mid2.loc[df_mid2['additional_payer_specific_notes'].str.startswith('EAPG-'), 'eapg'] = df_mid2[df_mid2['additional_payer_specific_notes'].str.startswith('EAPG-')]['additional_payer_specific_notes'].str.replace('EAPG-', '')
df_mid2.loc[df_mid2['additional_payer_specific_notes'] == '', 'additional_payer_specific_notes'] = None
df_mid2
```

```python
df_mid2['additional_generic_notes'] = df_mid2['contracting_method']
df_mid2['contracting_method'] = 'other'
df_mid2.loc[df_mid2['additional_generic_notes'] == 'Per Diem', 'contracting_method'] = 'per diem'
df_mid2.loc[df_mid2['additional_generic_notes'] == 'Per Case', 'contracting_method'] = 'case rate'
df_mid2.loc[df_mid2['additional_generic_notes'] == 'Straight Discount', 'contracting_method'] = 'percent of total billed charge'
df_mid2
```

```python
df_mid2['standard_charge'] = df_mid2['standard_charge'].astype(str)
df_mid2.loc[df_mid2['standard_charge'].str.endswith('%'), 'standard_charge_percent'] = df_mid2[df_mid2['standard_charge'].str.endswith('%')]['standard_charge'].str.replace("%", "")
df_mid2.loc[df_mid2['standard_charge'].str.endswith('%'), 'standard_charge'] = None
df_mid2
```

```python
df_mid2['description'] = df_mid2['description'].str.strip()
df_mid2['payer_name'] = df_mid2['payer_name'].str.strip()
```

```python
df_mid2['hospital_id'] = ccn
df_mid2['local_code'] = None
df_mid2['apr_drg'] = None
df_mid2['modifiers'] = None
df_mid2['alt_hcpcs_cpt'] = None
df_mid2['thru'] = None
df_mid2['ndc'] = None
df_mid2['drug_hcpcs_multiplier'] = None
df_mid2['drug_quantity'] = None
df_mid2['drug_unit_of_measurement'] = None
df_mid2['drug_type_of_measurement'] = None
df_mid2['billing_class'] = None
df_mid2['plan_name'] = None
df_mid2
```

```python
df_out2 = pd.DataFrame(df_mid2[TARGET_COLUMNS])
df_out2
```

```python
df_out2.to_csv('rate_' + ccn + '_tbl2.csv', index=False)
```
